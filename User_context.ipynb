```python
forbidden_imgs = {}
class preproc:
  def __init__(self):
    self.tokens_lst = []

    self.english_words = set(words.words())

  def do_preproc(self, lst):
    lst = self.lcase(lst)

    lst = self.remove_hashtags_and_tags(lst)

    lst = self.remove_puncts(lst)

    is_english = self.remove_non_english(lst)

    if is_english:
      self.tokenize(lst)
      return self.tokens_lst
    return None
  
  def lcase(self, lst):
    lst = lst.lower()
    return lst
  
  def remove_non_english(self, lst):
    eng_count, total = 0, 0
    for word in lst.split(' '):
      if word in self.english_words:
        eng_count+=1
      total+=1
    if total == 0:
      return False
    if eng_count/total <= 0.3:
      return False
    return True

  def remove_hashtags_and_tags(self, lst):
    curstr = ""
    for word in lst.split(' '):

      if len(word) == 0 or word[0] == '@' or word[0] == '#':
        continue

      curstr += word + ' '

    lst = curstr
    return lst

  def remove_puncts(self, lst):
    for character in string.punctuation:
      lst = lst.replace(character, '')
    return lst
  
  def stem(self, lst):
    curstr = ''
    for word in lst.split(' '):
      curstr += self.stemmer.stem(word) + ' '
    lst = curstr
    return lst

  def remove_stopwords(self, lst):
    curstr = ''
    for word in lst.split(' '):
      if word not in self.stopwords:
        curstr += word + ' '
    lst = curstr
    return lst

  def tokenize(self, lst):
    curlst = []
    final_list = []
    curlst = word_tokenize(lst)
    for word in curlst:
      is_eng = True
      for character in word:
        if not(97 <= int(ord(character)) <= 122) and not(48<= int(ord(character)) <= 57):
          is_eng = False
          break
      
      if is_eng == True:
        final_list.append(word)
    final_list.append(word)
    self.tokens_lst = final_list

    

```

```python
class word:
  def __init__(self, value):
    self.val = value
    self.doc_set = None
    self.term_freq_dict = {}
  
```

```python
def update_dict(dictionary, tokens, doc_id):
  for token in tokens:
    curword = None
    if token in dictionary.keys():    
      curword = dictionary[token]
    else:
      curword = word(token)
      curword.doc_set = {doc_id}
      dictionary[token] = curword
    
    if doc_id not in curword.term_freq_dict.keys():
      curword.term_freq_dict[doc_id] = 1
    else:
      curword.term_freq_dict[doc_id] +=1
    curword.doc_set.add(doc_id)
def make_tf_idf(dictionary, total_docs):
  #given a word we know it's tf for a doc and df as well
  # calc tf_idf add them and do div by len(doc_list) 
  tf_idf_dict = {}
  for token in dictionary:

    curword = dictionary[token]
    inter_idf = (total_docs+1)/(len(curword.doc_set)+1)
    idf = math.log(inter_idf)

    tf_idf_sum = 0
    length = 0

    for doc_id in curword.term_freq_dict:
      tf_idf_sum += (1+math.log(curword.term_freq_dict[doc_id]))*idf
      length+=1

    tf_idf_dict[token] = tf_idf_sum/length
  return tf_idf_dict

```
